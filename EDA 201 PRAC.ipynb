```python
#FINDING OUTLIERS USING IQR
```

```python
def finding_outliers(x) : 
    Q1 = df[x].quantile (0.25)
    Q3 = df[x].quantile (0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    outliers = df[(df[x] < lower_bound) | (df[x] >upper_bound)]

    print ('')
    print (f"finding outliers for {x}")
    return outliers[x]
```

```python
column_list = ['bathrooms','bedrooms','price','square_feet']
```

```python
for x in column_list :
    outliers = finding_outliers(x)
    print(outliers)
```
